```python
import requests
from bs4 import BeautifulSoup
from time import sleep
import random
import re
import urllib
import lxml.html
import pandas as pd
```

```python
list1 = []
for page in range(21,48):
    sleep(random.randint(5, 10))
    connection = urllib.urlopen('https://jobscentral.com.sg/jobsearch?q=data&pg='+str(page))

    dom =  lxml.html.fromstring(connection.read())
    for url in dom.xpath('//a/@href'): # select the url in href for all a tags(links)
        list1.append(url)
print list1
```

```python
list2 = [job for job in list1 if '/job/' in job]
list3 = list(set(list2))
list3
```

```python
# declare a list for final output
titles = []
companies = []
descriptions = []
attributes = []

for link in list3:
    sleep(random.randint(3, 6))
    url = 'https://jobscentral.com.sg'+link.format(page)
    #print url

    # make request for that page
    r = requests.get(url,allow_redirects=False)
    
    # turn into a BeautifulSoup object
    soup = BeautifulSoup(r.text, 'lxml')
    #print soup
    
    # for job titles
    title = soup.find('h1', {'class': 'title'})
    if title is None:
        title = 'NA'
    else: 
        title = soup.find('h1', {'class': 'title'}).get_text() # this works
    titles.append(title)
    #print titles
    
    # for company names
    company = soup.find('h2', {'class': 'company'})
    if company is None:
        company = 'NA'
    else:
        company = soup.find('h2', {'class': 'company'}).get_text() # this works
    companies.append(company)
    
    # for job description
    desc = soup.find('div', {'class': 'job-description-container'})
    if desc is None:
        desc = 'NA'
    else: 
        desc = soup.find('div', {'class': 'job-description-container'}).get_text() # this works
    descriptions.append(desc)
          
    # for job categories
    try:
        jobcategory = soup.findAll('dd', {'class': 'col-xs-8 col-sm-9'}) # this works
    except Exception:
        pass
    jobattr = [attr.get_text() for attr in jobcategory]
    attributes.append(jobattr)
    
#     print title
#     print company
#     print desc
#     print jobcategory
#     print jobattr
```
